```python
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
def get_result_dict(file_name):
    with open(file_name, "r") as f:
        return json.loads(f.readline())

def show_mean_step_num_graph(file_name):

    result = get_result_dict(file_name)

    data = np.array([[episode["step_num"] for episode in test]for test in result])
    mean_action_step = data.mean(axis=0)
    plt.plot(mean_action_step)
```

```python
show_mean_step_num_graph(file_name = "./../data/test_result_5x5_forget.txt")
```

```python
show_mean_step_num_graph(file_name = "./../data/test_result_5x5_no_forget.txt")
```

```python
plt.subplot(221)
show_mean_step_num_graph(file_name = "./../data/test_result_10x10_no_forget.txt")

plt.subplot(222)
show_mean_step_num_graph(file_name = "./../data/test_result_10x10_forget.txt")

plt.subplot(223)
show_mean_step_num_graph(file_name = "./../data/test_result_5x5_no_forget.txt")

plt.subplot(224)
show_mean_step_num_graph(file_name = "./../data/test_result_5x5_forget.txt")
```

```python
def get_result_dict(file_name):
    with open(file_name, "r") as f:
        return json.loads(f.readline())

def f(file_name):

    result = get_result_dict(file_name)

    data = np.array([[episode["step_num"] for episode in test]for test in result])
    return data.mean()

for threshold in np.arange(1, 2.300001, 0.1):
    for ratio in np.arange(0.1, 1.00001, 0.1):
        threshold = np.round(threshold, 2)
        ratio = np.round(ratio, 2)
        print(threshold, ratio, ": ", f(f"./../data/tuning/5x5_({threshold}, {ratio}).txt"))
```
