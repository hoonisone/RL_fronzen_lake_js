```python
import numpy as np
import json
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
def get_data(save_path):

    with open(save_path, "r") as f:
        return json.loads(f.readline())
    
def get_acc(x):
    x = [v["reward"] for v in x]
    for i in range(1, len(x)):
        x[i] += x[i-1]
    return x

plt.plot(get_acc(get_data(Path("./result/planning.json"))))
plt.plot(get_acc(get_data(Path("./result/no_planning.json"))))
```
