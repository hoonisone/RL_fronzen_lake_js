```python
import json
import numpy as np

from matplotlib import pyplot as plt
```

```python
with open('./data/baseline_result.json') as f:
    baseline_result = json.load(f)
    
with open('./data/forget_result.json') as f:
    forget_result = json.load(f)
```

```python
baseline_result = np.array(baseline_result).mean(0)
forget_result = np.array(forget_result).mean(0)
```

```python
plt.plot(baseline_result, label = "forget x")
plt.plot(forget_result, label = "forget o")
plt.legend()
```
